```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("C:/Users/student/Pictures/Saved Pictures/logo.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')  
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 50, 200)
plt.imshow(edges,cmap='grey')
```

```python
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 68, minLineLength=50, maxLineGap=5)

for line in lines:
   x1, y1, x2, y2 = line[0]
   cv2.line(image, (x1, y1), (x2, y2), (255, 0, 0), 3)
print("Line Detection using Hough Transform")
plt.imshow(image)
```

```python

image = cv2.imread("C:/Users/student/Pictures/Saved Pictures/logo.jpg", cv2.IMREAD_GRAYSCALE)


if image is None:
    print("Error: Image not found.")
else:

    image_float = np.float32(image)
    harris_response = cv2.cornerHarris(image_float, 2, 3, 0.04)


    harris_response = cv2.dilate(harris_response, None)  
    
    color_image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
    
    
    color_image[harris_response > 0.01 * harris_response.max()] = [0, 0, 255]  

    plt.imshow(color_image)
    plt.axis('off')  # Hide the axe
    plt.show()
```
